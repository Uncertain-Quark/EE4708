## 1. SVM (Linearly Seperable Data)

### You are given a two dimensional dataset (SVM_Q1) with 4 classes and you have to use SVM for classification.

### Import the required libraries 

```python
# Imports library
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.metrics import plot_confusion_matrix
from sklearn.model_selection import train_test_split
```

### The code for loading data is given below.

```python
# load data (do not change it)
import numpy as np
X_train, X_test, y_train, y_test = np.load('./SVM_Q1.npy', allow_pickle=True)
```

### 1.a) Write code to visualize the training data

```python
# Generate scatter plot for training data 
plt.scatter(X_train[y_train==0,0],X_train[y_train==0,1],c='r')
plt.scatter(X_train[y_train==1,0],X_train[y_train==1,1],c='g')
plt.scatter(X_train[y_train==2,0],X_train[y_train==2,1],c='b')
plt.scatter(X_train[y_train==3,0],X_train[y_train==3,1],c='c')
plt.grid()
plt.show()
```

### 1.b) Write the code for SVM classifier (you can write code from scratch or use sklrean)

```python
# SVM classifier code here
clf_svc = SVC(kernel = 'linear')
clf_svc.fit(X_train,y_train)
```

### 1.c) Generate the confusion matrix for the classifer on the test data

```python
# Generate confusion matrix
plot_confusion_matrix(clf_svc , X_test , y_test)
```

### 1.d) Visualize the support vectors and decision boundary

```python
# Code here

######### Visualizinf support vectors ##########
plt.scatter(X_train[:,0], X_train[:,1])
plt.scatter(clf_svc.support_vectors_[:,0],clf_svc.support_vectors_[:,1],c = 'red',label = 'support vectors')
plt.grid()
plt.title('Visualising the support vectors')
plt.legend()
plt.show()
######## Decision Boundary #############
x , y =  np.linspace(-5 , 9, 200) , np.linspace(-4 , 9 , 200)
n_x , n_y = np.meshgrid(x , y)

pred = clf_svc.predict(np.c_[n_x.ravel(), n_y.ravel()])
plt.contourf(n_x , n_y , pred.reshape(n_x.shape) )
plt.title('Decision Boundary for the linearly seperable data')
plt.show()
```

### 1.e) Report the accuracy obtained 

```python
# Code for Accuracy
print('Accuracy on the test set is :',clf_svc.score(X_test , y_test))
```

## 2. SVM (Non-linearly separable Data)

### You are given a two dimensional dataset with 2 classes as loaded below and you have to use SVM for classification.

```python
# generating data (do not change it)
from sklearn.datasets import make_circles 
X, Y = make_circles(n_samples = 3000, noise = 0.04) 
```

### 2.a) Use train_test_split function in sklearn to split the dataset into train and test sets in the ratio of 80:20

```python
# Code here
X_train , X_test , y_train , y_test = train_test_split(X , Y , test_size = 0.2)
```

### 2.b) Write the code for SVM classifier with polynomial kernel (you can write code from scratch or use sklrean)

Hint: Use kernel of degree 8.

```python
# SVM code here
clf_svc_non_linear = SVC(kernel = 'poly' , degree = 8)
clf_svc_non_linear.fit(X_train , y_train)
```

### 2.c) Generate the confusion matrix for the classifer on the test data

```python
# Generate confusion matrix
plot_confusion_matrix(clf_svc_non_linear , X_test , y_test)
```

### 2.d) Visualize the support vectors and plot the decision boundary

```python
# Visualize support vectors
######### Visualizinf support vectors ##########
plt.scatter(X_train[:,0], X_train[:,1])
plt.scatter(clf_svc_non_linear.support_vectors_[:,0],clf_svc_non_linear.support_vectors_[:,1],c = 'red',label = 'support vectors')
plt.grid()
plt.title('Visualising the support vectors')
plt.legend()
plt.show()
######## Decision Boundary #############
x , y =  np.linspace(-1 , 1, 200) , np.linspace(-1 , 1 , 200)
n_x , n_y = np.meshgrid(x , y)

pred = clf_svc_non_linear.predict(np.c_[n_x.ravel(), n_y.ravel()])
plt.contourf(n_x , n_y , pred.reshape(n_x.shape) )
plt.title('Decision Boundary for the non linearly seperable data')
plt.show()
```

### 2.e) Report the accuracy obtained 

```python
# Accuracy code
print('Accuracy on the test set is :',clf_svc_non_linear.score(X_test , y_test))
```

### 2.f) Repeat the above steps with sigmoid kernel (you can write code from scratch or use sklearn). You can tune the parameter to obtain better accuracy on test set.

```python
#Code here
clf_svc_sigmoid = SVC(kernel = 'sigmoid',gamma = 1e0)
clf_svc_sigmoid.fit(X_train , y_train)

plot_confusion_matrix(clf_svc_sigmoid , X_test , y_test)
plt.show()
######### Visualizinf support vectors ##########
plt.scatter(X_train[:,0], X_train[:,1])
plt.scatter(clf_svc_sigmoid.support_vectors_[:,0],clf_svc_sigmoid.support_vectors_[:,1],c = 'red',label = 'support vectors')
plt.grid()
plt.title('Visualising the support vectors')
plt.legend()
plt.show()
######## Decision Boundary #############
x , y =  np.linspace(-1 , 1, 200) , np.linspace(-1 , 1 , 200)
n_x , n_y = np.meshgrid(x , y)

pred = clf_svc_sigmoid.predict(np.c_[n_x.ravel(), n_y.ravel()])
plt.contourf(n_x , n_y , pred.reshape(n_x.shape) )
plt.title('Decision Boundary for the non linearly seperable data')
plt.show()

print('Accuracy on the test set is :',clf_svc_sigmoid.score(X_test , y_test))
```

### 2.g) Reapeat the steps 2.c to 2.g with  radial basis function kernel (you can write code from scratch or use sklrean). You can tune the parameter to obtain better accuracy on test set.

```python
# Code here
clf_svc_rbf = SVC(kernel = 'rbf')
clf_svc_rbf.fit(X_train , y_train)

plot_confusion_matrix(clf_svc_rbf , X_test , y_test)
plt.show()
######### Visualizinf support vectors ##########
plt.scatter(X_train[:,0], X_train[:,1])
plt.scatter(clf_svc_rbf.support_vectors_[:,0],clf_svc_rbf.support_vectors_[:,1],c = 'red',label = 'support vectors')
plt.grid()
plt.title('Visualising the support vectors')
plt.legend()
plt.show()
######## Decision Boundary #############
x , y =  np.linspace(-1 , 1, 200) , np.linspace(-1 , 1 , 200)
n_x , n_y = np.meshgrid(x , y)

pred = clf_svc_rbf.predict(np.c_[n_x.ravel(), n_y.ravel()])
plt.contourf(n_x , n_y , pred.reshape(n_x.shape) )
plt.title('Decision Boundary for the non linearly seperable data')
plt.show()

print('Accuracy on the test set is :',clf_svc_rbf.score(X_test , y_test))
```

### 2.h) For which kernal in SVM, you have obtained the highest and lowest  accuracy? What do you think the reason for it? 

For both polynomial and RBF kernel, we get equally good accuracy. However, the sigmoid kernel doesn't produce good accuracy on the test dataset. This is because the sigmoid kernel cannot produce closed shape forms such as is in case of data. However, in case of polynomial and RBF kernels, they are able to produce close shape forms as the decision boundary owing to which they demarcate the classes properly giving a better accuracy and model.

Highest accuracy : RBF , polynomial
